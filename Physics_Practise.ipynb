```python
import math as mt

def chk_prime(num):
  if(num == 1 or num == 0):
    print(str(num)+" is neither prime nor composite",end="\n\n");
  else :
    stpr=0;
    sq = mt.floor(mt.sqrt(num));
    for i in range(2,sq+1) :
      if(num%i == 0):
        stpr=1
        break
    if(stpr):
      print(str(num)+" is a composite number! ",end="\n\n");
    else:
      print(str(num)+" is a prime number! ",end="\n\n");

while(1) :
  a = str(input("Tell a number (TYPE EXIT TO LEAVE): "))
  if(a.lower() == "exit") :
    print("Exiting The Program !!");
    break;
  else:
    chk_prime(int(a));
```

```python
import fact as ft

def expo(num):
  exp=1
  for i in range(1,num+1):
    f=(num**i)/ft.fact(i)
    exp+=f
  return exp

a = int(input("Tell the value : "))
print(expo(a))
```

```python
def fibo(num):
  if(num == 0 or num == 1):
    return num
  else:
    return fibo(num-1) + fibo(num-2)

print(fibo(5))
```

```python
import matplotlib.pyplot as plt
import math as mt
import numpy as np

t = np.linspace(0, 2*np.pi ,1000)

x_erth = np.cos(t)
y_erth = np.sin(t)

x_mn = x_erth + .1*np.cos(14*t)
y_mn = y_erth + .1*np.sin(14*t)

x_ch = x_mn + 0.05*np.cos(90*t)
y_ch = y_mn + 0.05*np.sin(90*t)

plt.plot(x_erth, y_erth, label = 'Earth\'s orbit around the Sun')
plt.plot(x_mn, y_mn, label = 'Moon\'s orbit around the Sun')
plt.plot(x_ch, y_ch, label = 'Chandrayan\'s orbit around the Sun')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Chandrayan's Path with respect to moon and moon's path with respect to Earth")
plt.legend()

plt.grid(True)
plt.show()
```
